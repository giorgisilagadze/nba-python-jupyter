```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('all_seasons.csv')
df
```

```python
## გამოიტანთ ცხრილის სასურველ სტრიქონებს და სვეტებს,
selected_data = df[['player_name', 'team_abbreviation', 'pts', 'season', "age", "college", "draft_year", "pts", "reb", "ast"]].head(100)
selected_data
```

```python
## დაუნიშნეთ ინდექსირება ცხრილის კონკრეტული სვეტის მიმართ;
indexed_data = selected_data.set_index('player_name')
indexed_data
```

```python
## შექმნით 2 პარამეტრზე დამოკიდებულ ფილტრს. დაბეჭდეთ შესაბამისი ცხრილი.
filtered_data = df[(df['team_abbreviation'] == 'CLE') & (df['pts'] > 10)]
filtered_data
```

```python
## დაასორტირეთ ცხრილი 2 პარამეტრის გამოყენებით
sorted_data = filtered_data.sort_values(by=['pts', 'age'], ascending=[False, True])
sorted_data
```

```python
## გამოიყენეთ კონკრეტული სვეტის მნიშვნელობისთვის სტატისტიკური ფუნქციები (mean, standard, deviation, median, min, max).
stats = {
    'Mean': sorted_data['pts'].mean(),
    'Standard Deviation': sorted_data['pts'].std(),
    'Median': sorted_data['pts'].median(),
    'Min': sorted_data['pts'].min(),
    'Max': sorted_data['pts'].max()
}
stats
```

```python
## Numpy და Matplotlib-ის ბიბლიოთეკების გამოყენებით ააგეთ 2 სხვადასხვა ტიპის გრაფიკი (მაგ. Bar და ხაზოვანი დიაგრამები).
top_10_players = sorted_data[['player_name', 'pts']].drop_duplicates(subset=['player_name']).head(10)
top_10_players.plot(x='player_name', y='pts', kind='bar', title='ტოპ 10 მოთამაშე უმაღლესი საშუალო ქულით', figsize=(12, 6))
plt.ylabel('საშუალო ქულა')
plt.show()
```

```python
season_avg_pts = sorted_data.groupby('season')['pts'].mean()
season_avg_pts.plot(kind='line', title='სეზონების განმავლობაში CLE-ის მოთამაშეების მიერ მოპოვებული საშუალო ქულების ტენდენცია', figsize=(12, 6))
plt.ylabel('საშუალო ქულა')
plt.xlabel('სეზონი')
plt.show()
```
